# Pythagorean Expectation and its Application in Sports

Pythagorean Expectation is a statistical concept originally developed by baseball analyst Bill James. It is named after the Greek mathematician Pythagoras because it utilizes a formula derived from the Pythagorean theorem. While the concept was initially applied to baseball, it has also found applications in other sports.

The Pythagorean Expectation formula in sports estimates a team's winning percentage based on their runs (or points) scored and allowed. The formula is as follows:

Winning Percentage = (Runs (or Points) Scored) ^ exponent / ((Runs (or Points) Scored) ^ exponent + (Runs (or Points) Allowed) ^ exponent)

In this formula, the exponent is a constant value that varies depending on the sport and league being analyzed. It is usually set between 1.8 and 2.0 for baseball, while other sports may have different values.

The Pythagorean Expectation formula assumes that a team's performance is largely determined by their ability to score runs or points and prevent their opponents from doing the same. It suggests that a team's win-loss record should be correlated with their scoring differential. The larger the scoring differential, the higher the expected winning percentage.

By applying the Pythagorean Expectation to sports, analysts can assess the performance of teams beyond just their win-loss record. It helps identify teams that may be overperforming or underperforming based on their actual scoring differential. For example, if a team has a higher winning percentage than what is predicted by the Pythagorean Expectation, it may suggest that they have been winning close games by small margins and might be due for a regression. Conversely, a team with a lower winning percentage than expected may have experienced bad luck or performed poorly in close games.

The Pythagorean Expectation has been used in various sports, including basketball, football, and soccer, to evaluate team performance, assess the strength of a team's offense and defense, and predict future outcomes. It provides a useful tool for understanding the underlying factors that contribute to a team's success and can help identify teams that may be overachieving or underachieving based on their statistical performance.

# Pythagorean Expectation and the NBA

This is an example to showcase how pythogorean expectation can be applied to NBA. This 

```python
# Load the packages

import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data and see what it looks like

NBA = pd.read_csv('../../Data/Week 1/NBA_Games.csv')

NBA.head()
```

```python
# The data consists of games played between 2013 and 2019. 

# The season is identified by the column SEASON_ID
# Regular season game has 2 as prefix
# We can use the command ".describe()" to obtain descriptive statistics for our variables.

NBA_df = NBA[NBA.SEASON_ID == 22018]
NBA_df.describe()
```

```python
NBA['SEASON_ID'].unique()
```

```python
# We can list all the variable names

print(NBA_df.columns.tolist())
```

```python
# Many datasets contain missing variables. Missing variables in a column will usually cause operations to fail. 
# The command ".dropna()" will eliminate missing variables. 
# Compare the counts of variables below after the .dropna() below to the counts in the cell above.

NBA_df = NBA_df.dropna()
NBA_df.describe()
```

```python
# The game result is the column labeled 'WL'. We create a variable which has a value of '1' if the team won, and zero if it lost.
# This type of variable, where a condition (here winning) is either true (1) or not true (0) is called a "dummy variable".
# We will encounter them frequently.

NBA_df['result'] = np.where(NBA_df['WL']== 'W',1,0)
NBA_df.describe()
```

```python
# For the Pythagorean Expectation we need only the result, points scored (PTS) and point conceded (PTSAGN).

NBA_teams = NBA_df.groupby('TEAM_NAME')[['result','PTS','PTSAGN']].sum().reset_index()
NBA_teams
```

```python
# So now we can create the value for win percentage for each team in the 82 game season, and the Pythagorean Expectation.

NBA_teams['wpc'] = NBA_teams['result']/82
NBA_teams['pyth'] = NBA_teams['PTS']**2/(NBA_teams['PTS']**2 + NBA_teams['PTSAGN']**2)
NBA_teams
```

```python
# We now plot the data. 
sns.relplot(x="pyth", y="wpc", data = NBA_teams)
```

```python
# Finally we run the regression: wpc = Intercept + coef x pyth
# The coefficient of the variable pyth is strongly significant, and the R-Squared of the regression is close to 100%.

pyth_lm = smf.ols(formula = 'wpc ~ pyth', data=NBA_teams).fit()
pyth_lm.summary()
```

# Conclusion

The regression output tells you many things about the fitted relationship between win percentage and the Pythagorean Expectation. Regression is a method for identifying an equation which best fits the data. In this case that relationship is 

wpc = Intercept + coef x pyth

You can see the value of Intercept is -2.7572 and coef is 6.5536. It's this latter value were interested in. It means that for every one unit increase in pyth, the value of wpc goes up by 6.5536.

Two other points to note: 

(i) The standard error (std err) gives us an idea of the precision of the estimate. The ratio of the coefficient (coef) to the standard error is called the t statistic (t) and its value informs us about statistical significance. This is illustrated by the p-value (P > |t|) - this is the probability that we would observe the value 6.5536 by chance, if the true value were really zero. The summary of the regression model shows a p value of 0.00 which lets to say that the model is highly significant. 

(ii) in the top right hand corner of the table is the R-squared. This statistic tells you the percentage of variation in the y-variable (wpc) which can be accounted for by the variation in the x variables (pyth). R-squared can be thought of as a percentage - here the Pythagorean Expectation can account for 94.3% of the variation in win percentage.
